# QUESTION 1: "I don't understand how to loop through a list and retrieve a value when the value I need to retrieve is actually inside a dictionary?"  

If you have a list in Python and the list values are all dictionaries, the basic way to loop through the list AND the dictionary entries would be to use for loops.

```python
# This is a list with 2 values, which are dictionaries and contain key:value pairs
my_group = [ 
             {'name':'Bill', 'gender':'male', 'age':45, 'profession':'assistant professor'}, 
             {'name': 'Sarah', 'gender':'female', 'age': 48, 'profession':'associate professor'}
           ]

# Here we start a for loop and assign the variable 'i' to hold each value from the list 
# as we iterate in the loop
for i in my_group:
    # If we know what the keys of the dictionary are, we could just do this!
    print("This person's name is: {}".format(i['name']))
    print("This person's gender is: {}".format(i['gender']))
    print("This person's age is: {}".format(i['age']))
    print("This person's profession is: {}".format(i['gender']))
    print("\n")






#########################################################################


print("*"*30)
print("\n"*5)


# MY LIST WITH NESTED DICTIONARIES, NESTED LIST, and Values
another_group = [
        {'name':'Bill', 'gender':'male', 'age':45, 'profession':'assistant professor'},
        {'name':'Sarah', 'gender':'female', 'age': 48, 'profession':'associate professor'},
        ['foo', 'goo', 'boo'],
        'Apples'
        ]


# Our outer for loop is the same as above; we are just iterating through 
# each item in our list
for x in another_group:
    
    # Check to see if we have a dictionary
    if isinstance(x, dict):
        
        print("We found a nested dictionary:")
        
        # Here we execute another for loop to iterate through our key:value pairs
        # in the dictionary using our  .items() method
        for key, value in x.items(): 
            
            # We have the key AND value from each item in our dictionary
            print("\t{} : {}".format(key, value))
            
    # Check to see if we have a list
    elif isinstance(x, list):
        
        print("We found a nested list:")        
        
        # Loop through each item in our nested loop and print key and value
        # using .enumerate()
        for key, val in enumerate(x):
            
            print("\tWe have a list item -> {} : {}".format(key, val))
    
    # We only have a normal value in our list
    # so we just print it out
    else:
        
        print("The top-level value is: {}".format(x))
    
    print("\n")
```
